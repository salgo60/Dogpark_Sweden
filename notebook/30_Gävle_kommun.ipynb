```python
import geopandas as gpd
import folium
from IPython.display import display

# 1勇 L칛s in shapefilen
shapefile_path = "hundrastgardarGa팯vlekommun.zip"
gdf = gpd.read_file(f"zip://{shapefile_path}")

print("Kolumner:", gdf.columns.tolist())
print("Antal objekt:", len(gdf))
display(gdf)  # 九 visa hela tabellen

# 2勇 Konvertera till WGS84 (Folium kr칛ver lat/lon)
if gdf.crs is None or gdf.crs.to_epsg() != 4326:
    gdf = gdf.to_crs(epsg=4326)

# 3勇 Skapa Folium-karta
m = folium.Map(location=[60.6749, 17.1413], zoom_start=12, tiles="OpenStreetMap")

# 4勇 Lager f칬r ytor (polygoner)
polygon_layer = folium.FeatureGroup(name="游 Hundrastg친rdsytor", show=True)

for _, row in gdf.iterrows():
    name = row["Namn"]
    centroid = row.geometry.centroid
    lat, lon = centroid.y, centroid.x

    osm_url = f"https://www.openstreetmap.org/?mlat={lat:.6f}&mlon={lon:.6f}&zoom=17"
    popup_html = f"""
    <b>{name}</b><br>
    Lat: {lat:.6f}<br>
    Lon: {lon:.6f}<br>
    <a href="{osm_url}" target="_blank">游늸 Visa p친 OpenStreetMap</a>
    """

    folium.GeoJson(
        data=row["geometry"],
        name=name,
        tooltip=name,
        popup=folium.Popup(popup_html, max_width=300),
        style_function=lambda x: {
            "fillColor": "green",
            "color": "darkgreen",
            "weight": 2,
            "fillOpacity": 0.4,
        },
    ).add_to(polygon_layer)

polygon_layer.add_to(m)

# 5勇 Lager f칬r mark칬rer (centroid med ikon)
marker_layer = folium.FeatureGroup(name="游늸 Hundrastg친rdar (mark칬rer)", show=True)

for _, row in gdf.iterrows():
    name = row["Namn"]
    centroid = row.geometry.centroid
    lat, lon = centroid.y, centroid.x

    osm_url = f"https://www.openstreetmap.org/?mlat={lat:.6f}&mlon={lon:.6f}&zoom=17"
    popup_html = f"""
    <b>{name}</b><br>
    Lat: {lat:.6f}<br>
    Lon: {lon:.6f}<br>
    <a href="{osm_url}" target="_blank">游늸 Visa p친 OpenStreetMap</a>
    """

    folium.Marker(
        location=[lat, lon],
        popup=folium.Popup(popup_html, max_width=300),
        tooltip=name,
        icon=folium.Icon(color="darkblue", icon="paw", prefix="fa", icon_size=(45, 45))
    ).add_to(marker_layer)

marker_layer.add_to(m)

# 6勇 L칛gg till lagerkontroll
folium.LayerControl(collapsed=False).add_to(m)

# 7勇 Visa kartan
m
```
